```python
features = ['blueWardsPlaced', 'blueWardsDestroyed', 'blueFirstBlood',
       'blueKills', 'blueDeaths', 'blueAssists', 'blueEliteMonsters',
       'blueDragons', 'blueHeralds', 'blueTowersDestroyed', 'blueTotalGold',
       'blueTotalExperience', 'blueTotalMinionsKilled',
       'blueTotalJungleMinionsKilled', 'blueGoldDiff', 'blueExperienceDiff',
       'redWardsPlaced', 'redWardsDestroyed', 'redFirstBlood', 'redKills',
       'redDeaths', 'redAssists', 'redEliteMonsters', 'redDragons',
       'redHeralds', 'redTowersDestroyed', 'redTotalGold',
       'redTotalExperience', 'redTotalMinionsKilled',
       'redTotalJungleMinionsKilled', 'redGoldDiff', 'redExperienceDiff']
```

```python
dataset = lol.groupby('target')[features].mean()
print(dataset.head())
```

```python
#divide larger mean values by 1000 for better scaling/normalize in graph
dataset['blueTotalGold_1000'] = dataset['blueTotalGold'].div(1000)
dataset['blueTotalExperience_1000'] = dataset['blueTotalExperience'].div(1000)
dataset['blueGoldDiff_1000'] = dataset['blueGoldDiff'].div(1000)
dataset['blueExperienceDiff_1000'] = dataset['blueExperienceDiff'].div(1000)
dataset['redTotalGold_1000'] = dataset['redTotalGold'].div(1000)
dataset['redTotalExperience_1000'] = dataset['redTotalExperience'].div(1000)
dataset['redGoldDiff_1000'] = dataset['redGoldDiff'].div(1000)
dataset['redExperienceDiff_1000'] = dataset['redExperienceDiff'].div(1000)
print(dataset.head())
```

```python
dataset = dataset.drop(columns=['blueTotalGold', 'blueTotalExperience', 'blueGoldDiff', 'blueExperienceDiff',
                                'redTotalGold', 'redTotalExperience', 'redGoldDiff', 'redExperienceDiff'])
print(dataset.head())
```

```python
dataTypeSeries_2 = dataset.columns.to_series().groupby(dataset.dtypes).groups #set df.types as variable dataTypeSeries
print('Data type of each column of Dataframe :')
print(dataTypeSeries_2)
```

```python
#add new log features as features_2
features_2 = ['blueWardsPlaced', 'blueWardsDestroyed', 'blueFirstBlood', 'blueKills',
       'blueDeaths', 'blueAssists', 'blueEliteMonsters', 'blueDragons',
       'blueHeralds', 'blueTowersDestroyed', 'blueTotalMinionsKilled',
       'blueTotalJungleMinionsKilled', 'redWardsPlaced', 'redWardsDestroyed',
       'redFirstBlood', 'redKills', 'redDeaths', 'redAssists',
       'redEliteMonsters', 'redDragons', 'redHeralds', 'redTowersDestroyed',
       'redTotalMinionsKilled', 'redTotalJungleMinionsKilled',
       'blueTotalGold_1000', 'blueTotalExperience_1000', 'blueGoldDiff_1000',
       'blueExperienceDiff_1000', 'redTotalGold_1000',
       'redTotalExperience_1000', 'redGoldDiff_1000',
       'redExperienceDiff_1000']
```

```python
indx = np.arange(len(features_2))
print(indx)
```

```python
feature_label = np.arange(-2, 230, 20)
```

```python
blue_means = list(dataset.T[1])
red_means = list(dataset.T[0])
```

```python
print(dataset)
```

```python
bar_width = .35
fig, ax = plt.subplots(figsize=(30,10))
barBlue = ax.bar(indx - bar_width/2, blue_means, bar_width, label = 'Blue Team Wins', color='b')
barRed = ax.bar(indx + bar_width/2, red_means, bar_width, label = 'Red Team Wins', color='r')

ax.set_xticks(indx)
ax.set_xticklabels(features_2, rotation = 60, ha="right")

ax.set_yticks(feature_label)
ax.set_yticklabels(feature_label)
ax.set_title('Feature Means Comparison for Red and Blue wins', fontsize='large')
ax.set_xlabel('Features', fontsize='large')
ax.set_ylabel('Mean Values', fontsize='large')
ax.legend()

sns.set(style = 'ticks', font_scale = 1)

def insert_data_labels(bars):
    for bar in bars:
        bar_height = bar.get_height()
        ax.annotate('{0:.0f}'.format(bar.get_height()),
            xy=(bar.get_x() + bar.get_width()/2, bar_height),
            xytext=(0,3),
            textcoords='offset points',
            ha='center',
            va='bottom'
        )
        
insert_data_labels(barBlue)
insert_data_labels(barRed)


fig.tight_layout()
plt.show()
```
